# Pandas OPE Introduction

Pandas (panel data) is a library used to work with tabular data.  Most commonly used analytics libraries are compatible with Pandas Data Frames, making Pandas an essential tool to handling data. 

## Importing Pandas

Importing a library refers to adding functionality implemented in the library to your project.  
It is customary to import **Pandas** with the alias of **pd**, as shown below.  However, this is not the only way to import a library.

```python
import pandas as pd
```

## Creating Tabular Data
### Pandas Series
A Pandas Series is a column of values.

**NOTE: in Python _range(s,e,i)_ will create a list of numbers from _s_ to _e_ in increments of _i_.**

```python
my_series = pd.Series(range(10,30,3), name='this_is_my_series')
my_series
```

### Pandas Data Frames
A Pandas Data Frame consists of Pandas Series of the same length with the same row index.  
**NOTE: `tail()` will return the last 6 rows.  `tail(n)` will return the last n rows.  The same applies for `head()`, returning the first n rows.**

```python
my_dataframe = pd.DataFrame(data={'column_A': ['this','is','column','A','.'], 
                                  'column_B': ['I','am','column','B','!']})
my_dataframe.tail()
```

## Reading Tabular Data Files

**Pandas** includes a number of methods to import data from various sources, e.g. *csv*, *Excel*, *json*, *sql*, *pickle*, *parquet*, *etc.*  

**Note: Avoid using `read_pickle` to read pickled files from untrusted sources** 

For our examples we will show importing from *csv* and *Excel* files.  Other sources are very similar.  *Excel* requires other libraries e.g. *xlrd*.

### `read_csv` Function
Pandas *read_csv* has options for finer control, e.g. `parse_dates` will convert the column to *datetime* data type.  To view the available options, use "`help(pd.read_csv)`".

```python
# csv_df = pd.read_csv('../data/UNRATE.csv', parse_dates=['DATE'])
csv_df = pd.read_csv('https://raw.githubusercontent.com/emsanqui/Python_Training/master/data/UNRATE.csv', parse_dates=['DATE'])
csv_df.tail()
```

### `read_excel` Function

Pandas read_excel has options for finer control. To view the available options, use "`help(pd.read_excel)`".

```python
# xls_df = pd.read_csv('https://raw.githubusercontent.com/emsanqui/Python_Training/master/data/CPIAUCSL.csv', parse_dates=['observation_date'])
xls_df = pd.read_excel('https://raw.githubusercontent.com/emsanqui/Python_Training/master/data/CPIAUCSL.xls', skiprows=10, parse_dates=['observation_date'])
# xls_df = pd.read_excel('../data/CPIAUCSL.xls', skiprows=10, parse_dates=['observation_date'])
xls_df.tail()
```

## Selecting Columns

Use the `columns` attribute to list the columns in a Pandas data frame.

```python
my_dataframe.columns
```

### As a Series
To access a single Data Frame column as a Series, we have options.

```python
my_dataframe.column_A
```

```python
my_dataframe['column_A']
```

### As a Data Frame
One can also access a single column of a Data Frame as another Data Frame.  Note that when accessed as a Data Frame there is a column header.  Note the double `[]`

```python
my_dataframe[['column_A']]
```

### `iloc` Operator

`iloc` allows the user to select a column by column number.  Note that in Python indexes start with zero.   The `:` operator stands for **everything** in Python.

```python
my_dataframe.iloc[:,[1]]
```

## Selecting Rows

### `iloc` Operator
The `iloc` operator allows one to also select specific row numbers.  
In this example we select rows 1 through, but not including, 3.  Note that ranges do not include the end item. 

```python
my_dataframe.iloc[1:3]
```

We can also combine row and column selections.  
In this example we select rows 1 through 3 for columns **column_B**.

```python
my_dataframe.iloc[1:4,[1]]
```

### `loc` Operator
The `loc` operator is another option to select by label (columns or row name).  Note that with `loc` row 4 is included.

```python
my_dataframe.loc[1:4,['column_B']]
```

## Creating New Columns
We will use the Unemployment data for this example.

```python
csv_df.tail()
```

The most common way to add a column is declaring it as a new series in the data frame.  These can be calculations or values, which Pandas will broadcast. 

```python
csv_df['NEW COLUMN'] = "I'm new data"
csv_df['UNRATE_x_2'] = csv_df.UNRATE * 2
csv_df
```

One can also use the `assign()` Pandas function to insert new columns.  

```python
csv_df.assign(UNRATE_SQUARED=lambda x:x.UNRATE**2, ANOTHER_COLUMN='new col dat').tail()
```

If we did not specified the index at the time of import, we can create it with the `set_index()` function.

Note that if it is not reassigned to our `csv_df` demo data, the index will not stick.  One can pass use the `inplace` parameter to avoid having to reassign.

## Reshaping Data
For this section we will use Air Quality data for 2022 available from the EPA.

```python
co_df = pd.read_csv('https://raw.githubusercontent.com/emsanqui/Python_Training/master/data/ad_viz_plotval_data.csv', parse_dates=['Date'])
# co_df = pd.read_csv('../data/ad_viz_plotval_data.csv', parse_dates=['Date'])
co_df.head()
```

### `pivot_table` Function

Pandas includes functions to summarize data by groups, which will reshape the data frame, e.g. `pivot_table`, `pivot`, or `groupby`.

Below we calculate the total *CO Concentration* by *Date* by *County* using a pivot table.

```python
my_pt_df = co_df.pivot_table(index='Date', columns='COUNTY', values='Daily Max 8-hour CO Concentration', aggfunc=sum)
my_pt_df
```

### `melt` Function

Melt can be used to undo a pivot

```python
my_pt_df.reset_index().melt(id_vars=['Date'], var_name='COUNTY')
```

### `pivot` Function

I we do need any aggregation `pivot()` will do the trick.  Note that the `fillna()` function allows to replace any null values.

```python
(co_df[['COUNTY','Site ID','Site Name']]
 .drop_duplicates()
 .pivot(index='COUNTY', values='Site ID', columns='Site Name')
 .fillna('-'))
```

### `groupby` Function

We can use the `groupby()` function to apply functions by category

```python
co_df.groupby('COUNTY').Date.count()
```

## Plotting Data
For these examples we will use the 2022 Air Quality Index data from the EPA.  
Pandas built-in plotting back-end is Matplotlib, which supports various types of plots. 

```python
co_df.head()
```

### Time Series Plots
In the example below we plot the CO concentration time series data for California counties.

Note that in Python we can chain operators.  In this example we chained the `pivot_table` function output to the `plot` function.  If using multiple lines per operator, just wrap the statement in parenthesis.

```python
(co_df
 .pivot_table(index='Date', columns='COUNTY', values='Daily Max 8-hour CO Concentration', aggfunc=sum)
 .plot(figsize=(14,7))
);
```

### Histograms
Another type of plot supported include histograms.  Below we create a 21-bins histogram for the number of observations by `Site ID`.

```python
co_df['Site ID'].hist(figsize=(14,7), bins=21);
```

### Scatter Plots
We can also make a scatter plot of Longitude and Latitude

```python
co_df.plot(x='SITE_LONGITUDE', y='SITE_LATITUDE', alpha=0.5, kind='scatter', figsize=(14,7));
```

### Box Plots
The following is a box plot of the mean Air Quality Index by weekday.
Note that in Python we can concatenate strings with the '`+`' operator.

```python
(co_df
 .assign(DOW=lambda x:x.Date.dt.weekday.astype(str) + ' - ' + x.Date.dt.day_name())
 .pivot_table(index='COUNTY', columns='DOW', values='DAILY_AQI_VALUE', aggfunc='mean')
 .plot(kind='box', figsize=(14,7))
);
```

## Summary Statistics

### `describe` Function
The `describe()` function will provide summary statistics, ONLY for numerical columns.

```python
co_df.describe()
```

### `groupby` Function chained with a statistic function 

One can also get statistics for a single column by chaining.  Below we are calculating date counts by California county.

```python
co_df.groupby('COUNTY').Date.count()
```

### `agg` Function
We can also use the `agg()` function to get more than one statistic

```python
(co_df.groupby('COUNTY').agg({'DAILY_AQI_VALUE':['min','max','median','skew'], 'PERCENT_COMPLETE':['min','max']}))
```

The `value_counts()` function is another way to get counts by categories.  Below we count the number of observations by county.

```python
co_df.COUNTY.value_counts()
```
